```python
import re
def clean_tweets(text):
    text=re.sub("RT :","",text)
    text=re.sub("@[\w]*","",text)
    text=re.sub("https?://[A-Za-z0-9./]*","",text)
    text=re.sub("\n","",text)
    return text
```

```python
tweet_df['text']=tweet_df['text'].apply(lambda x: clean_tweets(x))
```
